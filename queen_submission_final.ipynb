```python
import random

#Function to generate the initial population

def initpop():
    return [[random.randint(1, n) for _ in range(n)] for _ in range(1000)]
    

#Function to score the population

def scorefit(chromosome):
    horizontal = sum([chromosome.count(queen)-1 for queen in chromosome])/2
    diagonal = 0

    n = len(chromosome)
    ld= [0] * 2*n
    rd = [0] * 2*n
    for i in range(n):
        ld[i + chromosome[i] - 1] += 1
        rd[len(chromosome) - i + chromosome[i] - 2] += 1

    diagonal = 0
    for i in range(2*n-1):
        counter = 0
        if ld[i] > 1:
            counter += ld[i]-1
        if rd[i] > 1:
            counter += rd[i]-1
        diagonal += counter / (n-abs(i-n+1))
    
    return int(maxscore - (horizontal + diagonal)) 

def pb(chromosome, scorefit):
    return scorefit(chromosome) / maxscore

def rd_select(population, probabilities):
    populationWithProbabilty = zip(population, probabilities)
    total = sum(w for c, w in populationWithProbabilty)
    r = random.uniform(0, total)
    upto = 0
    for c, w in zip(population, probabilities):
        if upto + w >= r:
            return c
        upto += w
#Function to do crossover

def reproduce(x, y):
    n = len(x)
    c = random.randint(0, n - 1)
    return x[0:c] + y[c:n]

#Function to mutate

def mutate(x):  
    n = len(x)
    c = random.randint(0, n - 1)
    m = random.randint(1, n)
    x[c] = m
    return x

#Function to create new generations

def genetic(population, scorefit):
    mutation = 0.01
    new_population = []
    probabilities = [pb(n, scorefit) for n in population]
    for i in range(len(population)):
        x = rd_select(population, probabilities) 
        y = rd_select(population, probabilities) #
        child = reproduce(x, y) 
        if random.random() < mutation:
            child = mutate(child)
        #print(child)
        new_population.append(child)
        if scorefit(child) == maxscore: break
    return new_population

# Main function to identify the right sequence
def main():
    
    population = initpop()
    generation = 1

    while not maxscore in [scorefit(chrom) for chrom in population]:
        print("Generation = {}".format(generation))
        population = genetic(population, scorefit)
        print("Best Fitness = {} ".format(max([scorefit(n) for n in population])))
        generation += 1
        #print(generation)
        
    final = []
    for chrom in population:
        if scorefit(chrom) == maxscore:
            final = chrom
    for i in range(len(chrom)):
        chrom[i] = chrom[i] - 1
    print("Final Solution")
    print(chrom)
    #print(generation)


#Triggering the main function with the n as 8
n=8
maxscore=(n*(n-1))/2
main()

```
